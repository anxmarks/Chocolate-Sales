```python
import pandas as pd
```

```python
df = pd.read_csv('../data/raw/Chocolate-Sales.csv')
```

```python
df.head()
```

```python
#remove $ and , from Amount and convert to float
df['Amount'] = df['Amount'].str.replace('$', '').str.replace(',', '').astype(float)
```

```python
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)
```

```python
df.to_csv('../data/processed/chocolate_sales.csv', index=False)
```
